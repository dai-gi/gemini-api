```python
class MyClass:
    """ A simple example class"""

    def __init__(self):
        self.name = "a"
        self.age = 12

    def getName(self):
        return self.name

    def setName(self, name):
        self.name = name

a = MyClass()
# a.setName("Tanaka")
print(a.name)
print(a.age)
```
